There does not seem to be any missing data right away, each variable appears three times (mean, se and worst) The data will be further explored and cleaned.

```python
# See how many people have been diagnosed with benevolent(0)/malignant(1) tumors.
df['diagnosis'].value_counts()
# The distribution does not seem to be extreme, there is about 1,5 benevolent for each malignant. 
```

```python
cols = ['diagnosis',
        'radius_mean', 
        'texture_mean', 
        'perimeter_mean', 
        'area_mean', 
        'smoothness_mean', 
        'compactness_mean', 
        'concavity_mean',
        'concave points_mean', 
        'symmetry_mean', 
        'fractal_dimension_mean']

sns.pairplot(data=df[cols], hue='diagnosis', palette='RdBu')  # palette='husl'
```

There clearly seems to a difference between the two groups (diagnosis), however when looking at the plots on the main diagonal (distribution of the data), there seems to be a lot of overlap between them.

```python
# Additionally, we create a correlation matrix to check how strongly components might be related.
# create a correlation matrix
c = df[cols].corr().abs()
sns.heatmap(c)
```

Besides the self-correlations, the strongest correlations appear to be radius and perimeter, radius and area. Considering they are so highly correlated, it might be best to remove them so they do not interfere with creating the model.

```python
cols = ['texture_mean', 
        'area_mean', 
        'smoothness_mean', 
        'compactness_mean', 
        'concavity_mean',
        'concave points_mean', 
        'symmetry_mean', 
        'fractal_dimension_mean']

y = np.array(df['diagnosis'])
X = np.array(df[cols])
X.shape
```

```python
# The data ranges are very large and first need to be normalized before using Scikit learn
from sklearn.preprocessing import StandardScaler

def normalize(X):
    scalar = StandardScaler()
    scalar = scalar.fit(X)
    X = scalar.transform(X)
    return X

X = normalize(X)
```

```python
# Split the data in train data and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
# Train model that fits a regression curve to the data
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
history = logreg.fit(X_train, y_train)
```

```python
# Evaluate trained model
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

y_pred = logreg.predict(X_test)
confusion_matrix = confusion_matrix(y_test, y_pred)
print ('=== [confusion matrix] ==='.center(54))
print(confusion_matrix)
print ('')
print ('=== [classification report] ==='.center(54))
print(classification_report(y_test, y_pred))
```

According to the confusion matrix, there seems to be 102 true positives for 4 false negatives. With 5 false positives with 60 true negatives. This translates to about 95% precision and recall for the model. This suggests the model fits the data very well.

```python
# Show the coefficients of the model
print(logreg.coef_)
```
